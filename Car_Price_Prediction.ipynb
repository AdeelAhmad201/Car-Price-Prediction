## Step 1: **Data Exploration**

```python
df = pd.read_csv("/content/CAR DETAILS FROM CAR DEKHO.csv")
```

# Step 2: **Feature Engineerig**

```python
fuel= pd.get_dummies(df['fuel'], drop_first = True)
seller_type= pd.get_dummies(df['seller_type'], drop_first = True)
transmission= pd.get_dummies(df['transmission'], drop_first = True)
owner = pd.get_dummies(df['owner'], drop_first = True)
```

```python
df = pd.concat([df,fuel,seller_type,transmission, owner], axis=1)
df
```

```python
df.info()
```

```python
boolean_columns = ['Diesel','Electric','LPG', 'Petrol', 'Individual', 'Trustmark Dealer', 'Manual',
                   'Second Owner','Test Drive Car']
for column in boolean_columns:
    df[column] = df[column].astype(int)
```

```python
df.drop(['fuel','seller_type', 'transmission', 'owner','name'], inplace = True, axis = 1)
```

```python
df.drop(['Fourth & Above Owner','Third Owner'], inplace = True, axis = 1)
```

```python
df
```

```python
x = df.drop('selling_price', axis = 1)
y= df['selling_price']
```

```python
from sklearn import preprocessing
pre_process =preprocessing.StandardScaler().fit(x)
x_transform = pre_process.fit_transform(x)
```

# Step 3: **Model Training & Evaluation**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x_transform, y, test_size= .20, random_state=101)
```

# **Gradient Boosting**

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
gb_regressor = GradientBoostingRegressor()
gb_regressor.fit(x_train,y_train)
y_pred_gb = gb_regressor.predict(x_test)
GBr= mean_squared_error(y_pred_gb,y_test)
print('Gradient Boosting regression :', GBr)
rmse = np.sqrt(GBr)
print("Root Mean Squared  :", rmse)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
rf_regressor = RandomForestRegressor()
rf_regressor.fit(x_train, y_train)
y_pred_rf =rf_regressor.predict(x_test)
RFr= mean_squared_error(y_pred_rf, y_test)
print("Random Forest Model :", RFr)
rmse = np.sqrt(RFr)
print("Root Mean Squared  :", rmse)
```
